```python
#finding outliers in registered column

sns.boxplot(vertical_concat_data['registered']).set(title="Box Plot of registered")
```

```python
#finding IQR (Interquartile Range)

IQR = 74-15 #Q3-Q1
print(IQR)

lower_limit = 15 - 1.5*IQR
upper_limit = 74 + 1.5*IQR

print(upper_limit)
print(lower_limit)
```

```python
#removing outliers or filtering

df_registered_filtered = vertical_concat_data['registered'][(vertical_concat_data['registered']>lower_limit) & (vertical_concat_data['registered']<upper_limit)]
print(df_registered_filtered)
```
